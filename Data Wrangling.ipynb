```python
sales = pd.read_csv(B)
sales['ordered_at'] = pd.to_datetime(sales['ordered_at'])
print(sales)
```

```python
print(sales.info())
```

```python
sales = sales.sort_values('ordered_at')
prices = prices.sort_values('updated_at')
```

```python
newtable = pd.merge_asof(sales,prices,left_on='ordered_at',right_on='updated_at',direction='nearest',by='product_id')
newtable
```

```python
newtable['final_price'] = np.where(newtable['ordered_at'] >= newtable['updated_at'],newtable['new_price'],newtable['old_price'])

newtable
```

```python
newtable['revenue'] = newtable['quantity_ordered'] * newtable['final_price']

newtable
```

```python
b1 = newtable.groupby(['product_id','final_price'])['revenue'].sum().reset_index()
b1
```
